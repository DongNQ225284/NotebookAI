## Zero-shot Learning (ZSL)

**Zero-shot learning** là kỹ thuật học máy cho phép mô hình phân loại các lớp **chưa từng thấy** trong quá trình huấn luyện.

### Đặc điểm:
- Dựa vào **mô tả ngữ nghĩa** (semantic embedding) như từ khóa, word vector, mô tả ngôn ngữ.
- Không cần dữ liệu huấn luyện cho mỗi class.

### Ứng dụng:
- CLIP (OpenAI): kết hợp ảnh và văn bản.
- Phân loại ảnh chưa có nhãn.
- NLP: dịch ngôn ngữ, phân loại chủ đề chưa có dữ liệu.

### Ví dụ: Sử dụng CLIP (OpenAI)

```python
import torch
import clip
from PIL import Image

# Tải mô hình CLIP
model, preprocess = clip.load("ViT-B/32")

# Chuẩn bị ảnh và văn bản
image = preprocess(Image.open("dog.jpg")).unsqueeze(0)
text = clip.tokenize(["a dog", "a cat", "a horse"])

# Trích đặc trưng và dự đoán
with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)
    logits_per_image, _ = model(image, text)
    probs = logits_per_image.softmax(dim=-1).numpy()

print("Label probs:", probs)
```
