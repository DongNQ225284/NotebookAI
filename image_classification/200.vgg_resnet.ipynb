## Mô hình VGG (Visual Geometry Group)

##### **Giới thiệu VGG16**
VGG16 là một kiến trúc mạng nơ-ron tích chập (CNN) được phát triển bởi nhóm Visual Geometry Group (VGG) tại Đại học Oxford. Mô hình này nổi bật trong cuộc thi ImageNet Challenge (ILSVRC) 2014, với độ chính xác cao và thiết kế đơn giản.

Tên gọi VGG16 xuất phát từ việc mô hình gồm 16 layers có trọng số học được: 13 convolutional layers và 3 fully connected layers.

##### **Đặc điểm chính của VGG16**
Sử dụng các kernel nhỏ 3×3, stride = 1 và padding = 1.

Tăng dần số lượng filters ở các layer sâu hơn (64 → 128 → 256 → 512).

Sử dụng max-pooling sau mỗi block để giảm chiều kích thước không gian.

Fully connected layers ở cuối trước khi phân loại bằng softmax.

Số lượng tham số lớn (~138 triệu), do đó rất nặng về bộ nhớ và tính toán.

##### **Kiến trúc VGG16**



##### **Ứng dụng của VGG16**
- Phân loại ảnh (image classification)
- Trích xuất đặc trưng (feature extraction)
- Chuyển học (transfer learning)
- Dùng làm backbone trong nhiều mô hình tiên tiến (Faster R-CNN, YOLO v1...)


```python
# type: ignore
from tensorflow.keras.applications import VGG16

model = VGG16(weights='imagenet', include_top=False, input_shape=(224,224,3))
model.summary()
```

## Mô hình ResNet (Residual Network)

**ResNet** sử dụng cơ chế **skip connection** (nối tắt) để khắc phục việc mạng càng sâu càng khó huấn luyện.

Block cơ bản:
$$
\text{Output} = \text{Activation}(F(x) + x)
$$

ResNet nổi bật:
- ResNet18, ResNet34, ResNet50, ResNet101...

```python
# type: ignore
from tensorflow.keras.applications import ResNet50

model = ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))
model.summary()
```
